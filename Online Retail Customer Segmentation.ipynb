```python
data['CustomerID'] = data['CustomerID'].astype(str)
```

## Data Preparation

### Customers will be classified based on the following factors:
- Count(C) = Number of transactions
- Recurrence(R) = Number of days since last purchase
- Amount(A) = Total amount or revenue generated

```python
data['Amount'] = data['Quantity']*data['UnitPrice']
data
```

```python
A_df = data.groupby('CustomerID')['Amount'].sum().reset_index()
A_df
```

```python
C_df = data.groupby('CustomerID')['InvoiceNo'].count().reset_index()
C_df.rename(columns={'InvoiceNo':'Count'}, inplace=True)
C_df
```

```python
AC_df = pd.merge(A_df, C_df, on='CustomerID', how='inner')
AC_df
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format= '%m/%d/%Y %H:%M')
data
```

```python
#Last Transaction date is equal to the max date
date_max = max(data['InvoiceDate'])
date_max
```

```python
#Days between max date and transaction date
data['Days'] = date_max - data['InvoiceDate']
data
```

```python
#Recurrence or number of days since last transaction
R_df = data.groupby('CustomerID')['Days'].min().reset_index()
R_df
```

```python
#Convert to number of days only
R_df['Days'] = R_df['Days'].dt.days
R_df.rename(columns={'Days':'Recurrence'}, inplace=True)
R_df
```

```python
ARC_df = pd.merge(AC_df, R_df, on='CustomerID', how='inner')
ARC_df
```

## Outlier Detection and Treatment

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=ARC_df)
plt.title('Outlier Distribution', fontsize=16)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Range', fontsize=12)
plt.savefig('P2-Outlier Distribution.png')
plt.show()
```

```python
#Removing outliers from Amount
A_Q1 = ARC_df.Amount.quantile(0.25)
A_Q3 = ARC_df.Amount.quantile(0.75)
A_IQR = A_Q3 - A_Q1
ARC_df = ARC_df[(ARC_df.Amount >= A_Q1 - 1.5*A_IQR) & (ARC_df.Amount <= A_Q3 + 1.5*A_IQR)]

#Removing outliers from Count
C_Q1 = ARC_df.Count.quantile(0.25)
C_Q3 = ARC_df.Count.quantile(0.75)
C_IQR = C_Q3 - C_Q1
ARC_df = ARC_df[(ARC_df.Count >= C_Q1 - 1.5*C_IQR) & (ARC_df.Count <= C_Q3 + 1.5*C_IQR)]

#Removing outliers from Recurrence
R_Q1 = ARC_df.Recurrence.quantile(0.25)
R_Q3 = ARC_df.Recurrence.quantile(0.75)
R_IQR = R_Q3 - R_Q1
ARC_df = ARC_df[(ARC_df.Recurrence >= R_Q1 - 1.5*R_IQR) & (ARC_df.Recurrence <= R_Q3 + 1.5*R_IQR)]
```

```python
ARC_df
```

## Data Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
ARC = ARC_df.copy()
ARC.drop('CustomerID', axis=1, inplace=True)
ARC
```

```python
scaler = StandardScaler()
ARC_scaled = scaler.fit_transform(ARC)
ARC_scaled
```

```python
ARC_scaled = pd.DataFrame(ARC_scaled)
ARC_scaled.columns = ['Amount','Count','Recurrence']
ARC_scaled
```

## Model Building

#### KMeans Clustering

```python
#Using elbow method to determine the optimal value of k
from sklearn.cluster import KMeans
ssd = {}
for n in range(1,11):
    KM = KMeans(n_clusters=n)
    KM.fit(ARC_scaled)
    ssd[n] = KM.inertia_

for key, val in ssd.items():
    print(f'{key}:{val}')
```

```python
ssd_l = list(ssd.values())

plt.figure(figsize=(15,8))
plt.plot(ssd_l)
plt.xlabel('Number of Clusters', fontsize=14)
plt.ylabel('Inertia', fontsize=14)
plt.title('Optimal Value of K via Elbow Method', fontsize=16)
plt.savefig('P2-Optimal Value of K in KMeans.png')
plt.show()
```

```python
#Final model with optimal value of K=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(ARC_scaled)
```

```python
kmeans.labels_
```

```python
ARC['Cluster'] = kmeans.labels_
ARC
```

```python
ARC.groupby('Cluster')['Amount'].sum()
```

```python
#Boxplot to visualize CLuster vs Amount

plt.figure(figsize=(15,8))
sns.boxplot(x=ARC['Cluster'], y=ARC['Amount'])
plt.xlabel('Cluster', fontsize=14)
plt.ylabel('Amount', fontsize=14)
plt.title('Cluster vs Amount', fontsize=16)
plt.savefig('P2-Cluster vs Amount.png')
plt.show()
```
